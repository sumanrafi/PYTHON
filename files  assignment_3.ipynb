```python
# Q2. What is exception handling in Python?

# solution
'''
    Exception handling in python is a mechanism to handle runtime error or exceptional
    condition that occur during program execution. It allow you to gracefully manage errors
    and prevent your program form crashing unexepectedly.

    1) Try Block - code that might raise an exception is placed inside the "try" block.
    2) Except Block - code to handle the exception is placed inside the "except" block.
    3) Finally Block - code that should run no matter what is placed inside the "finally" block.
    4) Else Block - code that should run if no exception is raised is placed inside the "else" block(optional).
    '''
```

```python
# Q9 What is the difference between multithreading and multiprocessing?

# solution
'''
Multithreading:-
1. multithreading uses multiple threads (lightweight sub-processes) within the same process to perform tasks concurrently.
2. all threads share the same memory space and resources within the process.
3. threads share memory, so it is more memory-efficient.
4. easier to communicate between threads since they share the same memory.

multiprocessing:-
1. multiprocessing uses multiple processes (independent execution units) to perform tasks concurrently.
2. each process has its own memory space and resources, making them isolated from each other.
3. each process has its own memory, which can lead to higher memory usage.
4. communication between processes requires inter-process communication (IPC) mechanisms like pipes or shared memory.
'''
```

```python
# Q14.What is the role of try and except in exception handling?

# solution
'''
Try Block :
            The try block contains the code that might raise an exception. It's the segment where you "try" to execute code that could potentially fail.
Except Block :
            The except block is used to handle the exception that is raised in the try block.You can specify different actions to take depending on the type of exception.
            If an exception occurs in the try block, the control is immediately transferred to the corresponding except block.

How They Work Together:
      1) Execution: The code inside the try block is executed line by line.
      2) Exception Occurs: If an exception occurs, the remaining code in the try block is skipped, and the control moves to the appropriate except block.
      3) Handling the Exception: The except block executes, handling the error in a defined way (e.g., printing an error message, logging the error, performing an alternative action).
      4) No Exception: If no exception occurs, the except block is skipped, and the code following the try...except structure continues to execute.
'''
```

```python
# Q18  How would you implement a basic logging setup that logs to a file with rotation after 1MB?

# solution
import logging
from logging.handlers import RotatingFileHandler
logger = logging.getLogger()
handler = RotatingFileHandler('error.log', maxBytes=1e6, backupCount=3)
logger.setLevel(logging.INFO)
formatter = logging.Formatter('%(asctime)s - %(message)s')
handler.setFormatter(formatter)
logger.addHandler(handler)
logger.info('This is an info message.')
logger.error('This is an error message.')
```

```python
# Q21 Write a Python program that reads a file and prints the number of occurrences of a specific word?

# solution
def count_word_in_file(file_path, target_word):
    try:
        with open(file_path, 'r') as file:
            count = 0
            for line in file:
                count += line.lower().split().count(target_word.lower())
        return count
    except FileNotFoundError:
        print(f"The file at {file_path} does not exist.")
        return 0

file_path = '/content/sample_data/practicle'
target_word = 'pw'

word_count = count_word_in_file(file_path, target_word)
print(f"The word '{target_word}' appears {word_count} times in the file.")
```

```python
# Q22 How can you check if a file is empty before attempting to read its contents?

# solution
import os

def is_file_empty(file_path):
    return os.path.getsize(file_path) == 0

def read_file(file_path):
    if is_file_empty(file_path):
        print(f"The file {file_path} is empty.")
    else:
        with open(file_path, 'r') as file:
            content = file.read()
            print("File content:")
            print(content)

file_path = '/content/sample_data/empty'

read_file(file_path)
```

```python
# Q23 Write a Python program that writes to a log file when an error occurs during file handling?

#solution
import logging

logging.basicConfig(
    filename='error.log',
    level=logging.ERROR,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def read_file(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            print(content)
    except Exception as e:
        logging.error(f"Error occurred while reading the file: {e}")

# Example usage
file_path = '/content/error.log'
read_file(file_path)

```
